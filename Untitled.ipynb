## Similarity Matrix
##### Create and save our similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity

df2_numpy = df2.values
similarity_matrix = cosine_similarity(df2_numpy)
sim_sorted = np.vstack([np.argsort(-row) for row in similarity_matrix])
np.save("similarity_mtrx", sim_sorted, allow_pickle=False)
sim_sorted.shape
```

```python
def similars(x):
    sim_ids = sim_sorted[x.name, 1:4]
    for i, item in enumerate(sim_ids):
        item = df.loc[item, 'brand'] + ' ' + df.loc[item, 'laptop_name']
        x[str(i+1)+'st similar'] = item
    return x
df_similarities = df.apply(similars, axis=1)
df_similarities.head()
```
